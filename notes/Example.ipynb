```python
import ast
```

```python
with open("raw.py", "r") as f:
    code = f.read()
```

```python
code_ast = ast.parse(code, mode="exec")
print(ast.walk(code_ast))
```

```python
for n, _ in enumerate(ast.walk(code_ast)):
    # print(n , _)
    if isinstance(_, ast.Assign):
        try:
            print(n, _, "lineno =",  _.lineno)
        except:
            pass
        try:
            print(_.targets[0].id)
        except AttributeError:
            print(_.targets[0].value.id)  
```
